```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    
    return L
```

```python
movies_df.iloc[0].genres
```

```python
movies_df['genres'] = movies_df['genres'].apply(convert)
movies_df['keywords'] = movies_df['keywords'].apply(convert)
```

```python
movies_df.iloc[0].genres
movies_df.iloc[0].keywords
```

```python
def convert2(obj):
    L=[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter ++1
        else:
            break
    return L
```

```python
movies_df['cast'] = movies_df['cast'].apply(convert2)
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
    
    return L
```

```python
movies_df['crew'] = movies_df['crew'].apply(fetch_director)
```

```python
movies_df.iloc[0].crew
```

```python
movies_df['overview'] = movies_df['overview'].apply(lambda x:x.split())
```

```python
movies_df['overview'][0]
```

```python
movies_df['genres'] = movies_df['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
movies_df['keywords'] = movies_df['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
movies_df['cast'] = movies_df['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
movies_df['crew'] = movies_df['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies_df.head()
```

```python
movies_df['tags'] = movies_df['overview'] + movies_df['genres'] + movies_df['keywords'] + movies_df['cast'] + movies_df['crew']
```

```python
movies_df.iloc[0].tags
```

```python
new_df = movies_df[['movie_id', 'title', 'tags']]
new_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv  = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
cv.fit_transform(new_df['tags']).toarray().shape
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
vector[0]
```

```python
len(cv.get_feature_names())
```

```python
import nltk
from nltk.stem.porter import  PorterStemmer
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(PorterStemmer().stem(i))
    
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vector)
```

```python
cosine_similarity(vector).shape
```

```python
sorted(list(enumerate(cosine_similarity(vector[0]))), reverse = True, key = lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = cosine_similarity(vector[movie_index])
    movies_list = sorted(list(enumerate(distances)), reverse = True, key = lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('The Dark Knight Rises')
```
